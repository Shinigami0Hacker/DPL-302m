```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
df = pd.read_csv("Dataset//depression_analysis_US.csv") 
```

```python
df
```

```python
year = df['year']
pop = df['Populations']
anx_rate = df['Anxiety rate']
anx = df['Anxiety case']

dep_rate = df['Depressive rate']
dep = df[' Depressive case ']
```

```python
plt.plot(year,anx_rate,label='Anxiety incidence',color='r')
plt.plot(year,dep_rate,label='Depression incidence',color='g')
plt.title("Mental health issues incidence in US")
plt.xlabel("Year")
plt.ylabel("Incidence")
plt.legend()
plt.savefig('Img//Mentalhealth_incidence.png')
plt.show()
```

```python
def millions_formatter(x, pos):
    return f'{int(x/1e6)}M'
formatter = FuncFormatter(millions_formatter)
plt.gca().yaxis.set_major_formatter(formatter)

plt.plot(year,dep,color='g',label="Depression disorder cases")
plt.plot(year,anx,color='r',label="Anxiety disorders cases")
plt.title("Mental Health cases in US")
plt.xlabel("Year")
plt.ylabel("Number of cases")
plt.legend()
plt.savefig('Img//Mentalhealth_cases.png')
plt.show()
```

```python
plt.plot(year,pop)
plt.gca().yaxis.set_major_formatter(formatter)
plt.title("US population")
plt.xlabel("Year")
plt.ylabel("Population")
plt.savefig("Img//US_population.png")
plt.show()
```
