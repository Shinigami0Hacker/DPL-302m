```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dataset//cleaned_text_emotion.csv")
df.dropna(subset=['text'],inplace=True)
df.drop_duplicates(subset=['text'],inplace=True)
```

```python
text = df['text']
text = list(text)
lengths = [len(sent) for sent in text]

def describe(arr:list):
    mean = np.mean(arr)
    median = np.median(arr)
    min_arr = np.min(arr)
    max_arr = np.max(arr)
    
    print(f"Mean: {mean:.2f}\nMedian: {median}\nMin: {min_arr}\nMax: {max_arr}")
    
    return {"mean":mean,
            "median":median,
            "min":min_arr,
            "max":max_arr}
    
value = describe(lengths)
```

```python
plt.hist(lengths,bins=1000,edgecolor='black')
plt.ylabel('Number of observations')
plt.xlabel('Max length')
plt.savefig("Img//length_dis.png")
plt.show()
```

```python
tdt = pd.read_csv("Dataset//model_evaluate.csv")
loss = tdt['loss']
f1 = tdt['f1']
model = tdt['model']
```

```python
plt.figure(figsize=(6, 8))
plt.scatter(f1,loss)
count = 1
for i,label in enumerate(model):
    plt.text(f1[i] - 0.003,loss[i] + 0.004,str(count)+"."+label)
    count +=1
plt.title("Compare performance")
plt.xlabel("F1 score")
plt.ylabel("Loss")
plt.savefig("Img//Compare_performance.png")
plt.show()
```
